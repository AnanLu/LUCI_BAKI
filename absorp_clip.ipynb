```python
from os import path
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
def absorp_clip(cube,abs_spec,abs_wnb):
    
    if cube == 'sn3':
        
        x_sn3 = np.where((abs_wnb>=6430) & (abs_wnb<=6790))
        lambda_sn3 = abs_wnb[x_sn3]
        amplitude = abs_spec[x_sn3]
        wavenumber = 10**8/lambda_sn3
        
    if cube == 'sn2':
        
        x_sn2 = np.where((abs_wnb>=4770) & (abs_wnb<=5100))
        lambda_sn2 = abs_wnb[x_sn2]
        amplitude = abs_spec[x_sn2]
        wavenumber = 10**8/lambda_sn2
     
    return wavenumber, amplitude
```

```python
abs_spec = fits.getdata('../NGC524/absorption_test1.fits')
abs_wnb = fits.getdata('../NGC524/absorption_wnb_test1.fits')
```

```python
wnb3, amp3 = absorp_clip('sn3',abs_spec,abs_wnb)
```

```python
plt.plot(wnb3,amp3)
plt.show()
```

```python
wnb2, amp2 = absorp_clip('sn2',abs_spec,abs_wnb)
```

```python
plt.plot(wnb2,amp2)
plt.show()
```
